## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import norm, uniform
import numpy as np
import matplotlib.pyplot as plt



#The normal distribution with mean = 0 and standard deviation = 1 is known as the standard normal distribution.

def dnorm(x, mean=0, std=1): # Point estimate (PDF) for normal distribution
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def pnorm(q, mean=0, std=1):#Cumulative distribution function (CDF) for normal distribution
    result = norm.cdf(q, loc=mean, scale=std)
    return result
    
def qnorm(p, mean=0, std=1):#Quantile function for normal distribution
    result = norm.ppf(p, loc=mean, scale=std)
    return result


def rnorm(n, mean=0, std=1):#Generate random variables from the normal distribution.
    result = norm.rvs(loc=mean, scale=std, size=n)
    return result


#standard uniform distribution : lower = 0 and upper = 1
#scale = upper - lower  # Scale is the width of the distribution

def dunif(x, lower=0, upper=1):#probability density function (PDF) at x
    result = uniform.pdf(x, loc=mean, scale=std)
    return result

def punif(q, lower=0, upper=1): #cumulative distribution function (CDF) up to q
    result = uniform.cdf(q, loc=lower, scale=upper-lower)
    return result

def qunif(p, lower=0, upper=1):#quantile function for a given probability p
    result = uniform.ppf(p, loc=lower, scale=upper-lower)
    return result

def runif(n, lower=0, upper=1):#n random samples from a uniform distribution
    result = uniform.rvs(loc=lower, scale=upper-lower, size=n)
    return result

np.random.seed(42)
mean = 0
std = 1


# Normal distribution
data = np.arange(-5, 5, 0.01)
pdf_normal = dnorm(data, mean=mean, std=std)
cdf_normal = pnorm(data, mean=mean, std=std)
ppf_normal = qnorm(data, mean=mean, std=std)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(data, pdf_normal, 'b-', label='PDF')
plt.title('Normal Distribution PDF')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(data, cdf_normal, 'g-.', label='CDF')
plt.title('Normal Distribution CDF')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(data,ppf_normal, 'c^', label='PPF')
plt.title('Normal Distribution PPF')
plt.legend()

plt.tight_layout()
plt.show()

lower = 0
upper = 1

# Uniform distribution
data = np.arange(-5, 5, 0.01)
pdf_uniform = dunif(data, lower=lower, upper=upper)
cdf_uniform = punif(data, lower=lower, upper=upper)
ppf_uniform = qunif(data, lower=lower, upper=upper)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(data, pdf_uniform, 'b--', label='PDF')
plt.title('Uniform Distribution PDF')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(data, cdf_uniform, 'r:', label='CDF')
plt.title('Uniform Distribution CDF')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(data,ppf_uniform, 'm-', label='PPF')
plt.title('Uniform Distribution PPF')
plt.legend()

plt.tight_layout()
plt.show()



```
